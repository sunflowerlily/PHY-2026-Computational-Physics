## 1. 运行矢量化模拟

我们将模拟 100 个粒子的随机运动。

```python
N = 100
steps = 200
dt = 0.1

print(f"Running simulation for {N} particles...")
try:
    # 调用你的矢量化函数
    final_pos = simulate_nbody(N, steps, dt)
    
    # 简单可视化最终位置
    if np.all(final_pos == 0):
        print("⚠️ 警告: 所有粒子都在原点，你可能还没实现代码！")
    else:
        plt.scatter(final_pos[:, 0], final_pos[:, 1], alpha=0.6, c='blue')
        plt.title(f"Final Positions of {N} Particles")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid(True, alpha=0.3)
        plt.axis('equal')
        plt.show()
        
except Exception as e:
    print(f"💥 运行报错: {e}")
```
