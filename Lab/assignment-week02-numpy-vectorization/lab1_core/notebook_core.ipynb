```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import time

# 导入你的代码模块
sys.path.append('src')
from gravity_slow import compute_potential_slow
from gravity_fast import compute_potential_vectorized

# 设置 Matplotlib 样式
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (8, 6)
```

## 1. 场景设置：随机分布的 N 体系统

我们要生成一个包含 $N=20$ 个随机质量点的系统，并在 $200 \times 200$ 的网格上计算引力势。

```python
# 1. 初始化参数
N = 20  # 质量点数量
resolution = 200 # 网格分辨率

# 2. 生成网格
x = np.linspace(-5, 5, resolution)
X, Y = np.meshgrid(x, x)

# 3. 生成随机质量点
np.random.seed(42)
mass_pos = np.random.uniform(-4, 4, size=(N, 2))
mass_val = np.random.uniform(1, 5, size=N)
```

## 2. 运行慢速版本 (Baseline)

```python
print("Running Slow Version...")
V_slow = compute_potential_slow(X, Y, mass_pos, mass_val)
```

## 3. 运行你的向量化版本 (Student Task)

请确保你已经完成了 `src/gravity_fast.py` 中的代码填空。

```python
print("Running Fast Version...")
start = time.time()
try:
    V_fast = compute_potential_vectorized(X, Y, mass_pos, mass_val)
    print(f"Fast Version Time: {time.time() - start:.4f} s")
except NotImplementedError as e:
    print(f"Error: {e}")
    # 如果未实现，使用零矩阵代替以免绘图报错
    V_fast = np.zeros_like(X)
```

## 4. 可视化引力势能场

这里我们将绘制等势面。注意观察：
1.  **奇点**：在质量点附近，势能应该急剧下降（趋向负无穷）。
2.  **纵横比**：必须设置 `ax.set_aspect('equal')`，否则空间会被扭曲。

```python
fig, ax = plt.subplots()

# 使用 contourf 填充颜色
# levels=40 让颜色过渡更平滑
# cmap='inferno' 是感知均匀的颜色映射，适合表示强度
c = ax.contourf(X, Y, V_fast, levels=40, cmap='inferno')
plt.colorbar(c, ax=ax, label='Potential $\Phi$')

# 标出质量点的位置
ax.scatter(mass_pos[:, 0], mass_pos[:, 1], c='cyan', s=mass_val*20, edgecolors='white', label='Masses')

# 关键：设置等比例尺
ax.set_aspect('equal')
ax.set_title(f"Gravitational Potential (N={N})")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

plt.show()
```
